```python
## jSON 
# pip install json 
# 1era forma de trabajar con json 
import json

datos = {}
datos['clientes'] = [] 

# Agregar los clientes 

datos['clientes'].append({
    'nombre':'Ana',
    'edad': 19
})

datos['clientes'].append({
    'nombre':'Maria',
    'edad': 25
})

datos['clientes'].append({
    'nombre':'Julian',
    'edad': 35
})

datos['clientes'].append({
    'nombre':'Edgardo',
    'edad': 29
})

print(datos)

file = open('datos.json', 'w')
json.dump(datos, file)
    

```

```python
# Leerlos 
file = open('datos.json')
data = json.load(file)

for cliente in data['clientes']: 
    print("Nombre: ", cliente['nombre'])
    print("Edad: ", cliente['edad'])


    
```

```python
## Trabajando con XML 
dataxml = pd.read_xml('countries.xml')
dataxml.shape
```

```python
import xml.etree.ElementTree as et

tree = et.parse('countries.xml')
data = tree.getroot()

for pais in data.findall('country'):
    continente = pais.find('region').text
    if continente != 'Europe': 
        data.remove(pais)
        
tree.write('europa.xml')
```
